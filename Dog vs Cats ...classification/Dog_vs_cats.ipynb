```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
from data_builder import Data_Builder
import os
import shutil
import pandas as pd
import cv2

```

```python
def check_file(file):
        if os.path.exists(file):
            print("File Exists")
            return True
        else:
            print("File Does not exist")
            return False

def Split(path ,train_size = 0.8, ):

        images = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
        images_train = []
        count = 0
        for img in images:
            if count ==  (int(len(images)*train_size)+1):
               break
            images_train.append(img)
            images.remove(img)
        images_test = images
        
        check1 = check_file(f"{path}\\Train")
        check2 = check_file(f"{path}\\Test")
        if check1&check2 == False:
            os.mkdir(f"{path}\\Train")
            os.mkdir(f"{path}\\Test")
            for img in images_train:
                old_image_path = f"{path}\\{img}"
                new_image_path = f"{path}\\Train\\"
                shutil.move(old_image_path, new_image_path)
            for img in images_test:
                old_image_path = f"{path}\\{img}"
                new_image_path = f"{path}\\Test\\"
                shutil.move(old_image_path, new_image_path)
```

```python
data_dir = 'E:\DATA SETS\CAT vs Dogs\PetImages'
Split(f"{data_dir}\\Cat", 0.8)
```

```python
test_cat_x = []
train_cat_x = []
test_cat_y = []
train_cat_y = []

for i in os.listdir(f"{data_dir}\\Cat\\Train"):
    image = cv2.imread(f"{data_dir}\\Cat\\Train\\{i}")
    train_cat_x.append(image)
    train_cat_y.append('cat')
for i in os.listdir(f"{data_dir}\\Cat\\Test"):
    image = cv2.imread(f"{data_dir}\\Cat\\Test\\{i}")
    test_cat_x.append(image)
    test_cat_y.append('cat')
```

```python
test_Dog_x = []
train_Dog_x = []
test_Dog_y = []
train_Dog_y = []

for i in os.listdir(f"{data_dir}\\Dog\\Train"):
    image = cv2.imread(f"{data_dir}\\Dog\\Train\\{i}")
    train_Dog_x.append(image)
    train_Dog_y.append('dog')
for i in os.listdir(f"{data_dir}\\Dog\\Test"):
    image = cv2.imread(f"{data_dir}\\Dog\\Test\\{i}")
    test_Dog_x.append(image)
    test_Dog_y.append('dog')
```

```python
import random as r
```

```python
x_train =list(train_cat_x + train_Dog_x)
x_test  =list(test_cat_x + test_Dog_x)
y_test  =list(test_cat_y + test_Dog_y)
y_train =list(train_cat_y + train_Dog_y)

```

```python
r.shuffle(x_train)
r.shuffle(x_test)
r.shuffle(y_train)
r.shuffle(y_test)
```
