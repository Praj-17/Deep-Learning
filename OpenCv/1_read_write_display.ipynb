## Reading writing and displaying images in using opencV


```python
import cv2
import numpy as np

img1 = cv2.imread("Gallery\\1.png")
img2 = cv2.imread("Gallery\\2.jpg")
img3 = cv2.imread("Gallery\\4.jpg")
img4 = cv2.imread("Gallery\\5jpg.jpg")
img5 = cv2.imread("Gallery\\3jpg.jpg")
```

```python
cv2.imshow("Window name img1",img1 )
cv2.imshow("Window name img2",img2 )
cv2.imshow("Window name img3",img3 )
cv2.imshow("Window name img4",img4 )
cv2.imshow("Window name img5",img5 )
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(img1)
print(img2)
print(img3)
print(img4)
print(img5)
```

### Let us now explore how the opencv library sees an image

```python
print("width of the img1", img1.shape[1], "pixels")
print("width of the img2", img2.shape[1], "pixels")
print("width of the img3", img3.shape[1], "pixels")
print("width of the img4", img4.shape[1], "pixels")
print("width of the img5", img5.shape[1], "pixels")
```
